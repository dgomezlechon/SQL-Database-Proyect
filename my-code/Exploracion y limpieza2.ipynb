Importamos un csv con las coordenadas en las que fue la premiere de cada una de las películas. Vamos a plasmar en un mapa y ver si fuesemos un actor particular en que ciudad nos convendría vivir para estar más cerca de esas premieres

```python
import re
```

```python
import folium
```

```python
def geomap(loc_clientes_csv, zoom_start=10.5):

    loc_clientes_csv.index=range(len(loc_clientes_csv)) 

    mapa=folium.Map(location=[40.463667, -3.74922],
                tiles='openstreetmap', zoom_start=zoom_start)

    for i in range(1, len(loc_clientes_csv)):

        folium.CircleMarker([float(loc_clientes_csv.latitude[i]), float(loc_clientes_csv.longitude[i])], popup=str(loc_clientes_csv.customer_id[i]), 
                             radius=1, icon=folium.Icon()).add_to(mapa)

    #mapa.save('images/geomap.html')
    
    
    return mapa
```

```python
mapa=geomap(loc_clientes_csv.loc[:1500])
mapa
```

```python
def geomap(shops_csv, zoom_start=11.5):

    shops_csv.index=range(len(shops_csv)) 

    mapa=folium.Map(location=[40.463667, -3.74922],
                tiles='openstreetmap', zoom_start=zoom_start)

    for i in range(1, len(shops_csv)):

        folium.Marker([float(shops_csv.latitude[i]), float(shops_csv.longitude[i])], popup=str(shops_csv.shop_id[i]), 
                             radius=1, icon=folium.Icon()).add_to(mapa)

    #mapa.save('images/geomap.html')
    
    
    return mapa
```
